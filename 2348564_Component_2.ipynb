created bins for Happiness score such as "low happiness","medium happiness","high happiness"

```python
bin_edges = [2.8,4,6,7.8]
bin_labels = ['low Happiness','medium Happiness', 'high Happiness']

df['Happiness Score Group'] = pd.cut(df['Happiness Score'], bins=bin_edges, labels=bin_labels, right=False)
```

```python
df
```

A bar chart showing the count of countries in each Happiness Score
group.

```python
df['Happiness Score Group'].value_counts().plot(kind='bar')
plt.title('Count of Countries in Each Happiness Score Group')
plt.xlabel('Happiness Score Group')
plt.ylabel('Count')
plt.show()
```

Calculate the covariance matrix for a subset of numerical columns in the dataset
(e.g Economy (GDP per Capita),Family,Health (Life Expectancy)).

```python
cols = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)']
df_subset = df[cols]


cov_matrix = df_subset.cov()
cov_matrix
```

A heatmap of the covariance matrix.

```python
sns.heatmap(cov_matrix, cmap='coolwarm')
plt.title('Covariance Matrix for Numerical Columns')
plt.show()
```
